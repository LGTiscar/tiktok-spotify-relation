```python
import pandas as pd
```

# TIKTOK DATA

```python
# let's load the data
top500_df = pd.read_csv('500top_tiktok.csv')
```

```python
# Clean unnecessary columns
top500music_df = top500_df[['id','musicMeta.musicId','musicMeta.musicName','musicMeta.musicAuthor','musicMeta.playUrl','webVideoUrl','diggCount',
'shareCount','playCount','commentCount']]

# Original sound synonims list to delete

os = ['original sound','suono originale','som original','son original','sonido original','orijinal ses','Originalton']

top500music_df_cleaned = top500music_df[~top500music_df['musicMeta.musicName'].isin(os)]
top500music_df_cleaned.sort_values('diggCount',ascending=False,inplace=True,ignore_index=True)

# Let's save it into a CSV file

top500music_df_cleaned.to_csv('top500tiktok.csv')
top500music_df_cleaned
```

# SPOTIFY DATA

## Top 50 Global Viral

```python
# Load the csv

df_spotify = pd.read_csv('50viral_28-10-2021.csv')
df_spotify
```

```python
df_spotify['song'] = df_spotify['Track Name'] + " | " + df_spotify['Artist']
```

```python
df_spotify
```

## Top 200 Global

```python
df_spotify_glob = pd.read_csv('200global_28-10-2021.csv')
df_spotify_glob['song'] = df_spotify_glob['Track Name'] + " | " + df_spotify_glob['Artist']
df_spotify_glob
```

# Check Matches

We are gonna check if there are coincidences between top50 globral viral, top 200 global spotify songs and Top500 TikTok songs

```python
top500=[]
for track in top500music_df_cleaned['musicMeta.musicName']:
    top500.append(track)
top500
```

```python
top50=[]
for track in df_spotify['Track Name']:
    top50.append(track)
top50

top200=[]
for track in df_spotify_glob['Track Name']:
    top200.append(track)
top200
```

## Top50 Matches

```python
count=0
matches=[]
for track in top50:
    if track in top500:
        matches.append(track)
        count +=1
print(f"There are {count} coincidences")
```

```python
# let's find out which ones are
print (*matches, sep="; ")
```

```python
index=[]
for match in matches:
    index.append(top500music_df_cleaned.index[top500music_df_cleaned['musicMeta.musicName'] == match].tolist()[0])
    
print(f'The position of top50 viral songs in top500 viral tiktoks are: {index}')
```

## Top 200 matches

```python
count200=0
matches200=[]
for track in top200:
    if track in top500:
        matches200.append(track)
        count200 +=1
print(f"There are {count200} coincidences")
```

```python
# Veamos cu√°les son:
print (*matches200, sep="; ")
```

```python
index200=[]
for match in matches200:
    index200.append(top500music_df_cleaned.index[top500music_df_cleaned['musicMeta.musicName'] == match].tolist()[0])
    
print(f'The position of top200 global songs in top500 viral tiktoks are: {index200}')
```

# RESULTS

```python
#Top 50 Results
results50_df = pd.DataFrame()
results50_df['Matches with top50 viral'] = matches
results50_df['Position in top500 tiktoks'] = index
results50_df.sort_values('Position in top500 tiktoks',inplace=True,ignore_index=True)

#Top 200 results
results200_df = pd.DataFrame()
results200_df['Matches with top200 global'] = matches200
results200_df['Position in top500 tiktoks'] = index200
results200_df.sort_values('Position in top500 tiktoks',inplace=True,ignore_index=True)
```

```python
print(f'The matches between the top50 global viral and the top200 viral tiktoks for the 28th of Octobre 2021 are:')
results50_df.to_csv('Results_top50.csv')
results50_df
```

```python
print(f'The matches between the top200 global and the top200 viral tiktoks for the 28th of Octobre 2021 are:')
results200_df.to_csv('Results_top200.csv')
results200_df
```
